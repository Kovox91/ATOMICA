# Data Preparation and Embedding

Start in the ATOMICA folder.
Activate the atomica virtual env:
```
source atomicaenv/lib/activate
```

### Step 1
Process the PeSTo outputs with the process_PeSTo_results.py file:
```
python -m data.process_PeSTo_results \
    --data_dir trial/pesto_out \
    --raw_data_dir trial/input \
    --output_dir trial/atomica_out \
    --prot_list trial/protein_list.txt \
    --b_factor_cutoff 0.6
```

### Step 2
Get the embeddings with get_embeddings.py
```
python -m get_embeddings \
    --model_config pretrained_model/pretrain_model_config.json \
    --model_weights pretrained_model/pretrain_model_weights.pt \
    --output_path trial/embeddings/embeddings.pkl \
    --data_path trial/atomica_out/pesto_60_ligand.pkl 
```
